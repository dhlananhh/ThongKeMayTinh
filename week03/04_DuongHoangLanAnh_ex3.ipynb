Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

````python
# Phát sinh mảng ngẫu nhiên gồm 100 phần tử
arr = np.random.randint(0, 1001, size=100)
````

````python
arr
````

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
# Hiển thị 10 phần tử đầu tiên
print(f"10 phần tử đầu tiên của mảng là: \n {arr[:10]}")

# Hiển thị 5 phần tử cuối cùng
print(f"5 phần tử cuối cùng của mảng là: \n {arr[-5:]}")

# Hiển thị các phần tử chia hết cho 5 và lớn hơn 100
condition = (arr % 5 == 0) & (arr > 100)
print(f"Các phần tử chia hết cho 5 và lớn hơn 100: \n {arr[condition]}")
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
filtered_numbers = [y for y in arr if y < 100]
print(f"Các phần tử có giá trị nhỏ hơn 100 là: \n {filtered_numbers}")
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
# Tính tổng mảng
sum = np.sum(arr)

# Tính giá trị trug bình
mean = np.mean(arr)

# Tính giá trị trung vị
median = np.median(arr)

# Tìm Q1, Q2, Q3
Q1 = np.percentile(arr, 25)
Q2 = np.percentile(arr, 50)
Q3 = np.percentile(arr, 75)

# Tìm phân vị 20 (percentile 20)
percentile_20 = np.percentile(arr, 20)

# In kết quả
print("Tổng mảng: ", sum)
print("Giá trị trung bình: ", mean)
print("Q1: ", Q1)
print("Q2: ", Q2)
print("Q3: ", Q3)
print("Phân vị 20: ", round(percentile_20, 2))
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
# Sử dụng 'np.random.choice' để chọn ngẫu nhiên 10 phần tử từ mảng trên 5 lần và hiển thị kết quả mỗi lần chọn

for i in range(5):
    random_selection = np.random.choice(arr, size=10, replace=False)
    print(f"Lựa chọn ngẫu nhiên lần {i + 1}: ", random_selection)
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
# Sử dụng 'np.bincount' để đếm tần suất xuất hiện của các phần tử có trong mảng
# Sử dụng 'argmax' để tìm ra phần tử xuất hiện nhiều nhất

most_common_element = np.bincount(arr).argmax()
print("Phần tử xuất hiện nhiều nhất trong mảng là: ", most_common_element)
````

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

````python
# Your code here
````

````python
# Phát sinh dữ liệu x từ 0 đến 2*pi
x = np.linspace(0, 2 * np.pi, 30)

# Tính giá trị y cho sin(x) và cos(x)
y = np.sin(x)
z = np.cos(x)

# Vẽ đồ thị với markers và đường nét liền
plt.plot(x, y, 'm.-', label='y = sin(x)')
plt.plot(x, z, 'b*--', label='z = cos(x)')

# Bật lưới
plt.grid(True)

# Đặt tên cho trục x và trục y
plt.xlabel('x')
plt.ylabel('y')

# Đặt tiêu đề cho đồ thị
plt.title('Đồ thị hàm sin và cos')

# Hiển thị chú thích
plt.legend()

# Hiển thị đồ thị
plt.show()
````

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
# Tính tổng tiền tip theo ngày
tip_by_day = tips_data.groupby('day')['tip'].sum().reset_index()

# Vẽ đồ thị
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='tip', data=tip_by_day)
plt.title('Tổng tiền tip theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng tiền tip')
plt.show()
````

### 2. Vẽ đồ thị thể hiện tổng tiền tip theo thời gian dùng bữa (time)

````python
# Tính tổng tiền tip theo thời gian dùng bữa
tip_by_time = tips_data.groupby('time')['tip'].sum().reset_index()

# Vẽ đồ thị
plt.figure(figsize=(8,6))
sns.barplot(x='time', y='tip', data=tip_by_time)
plt.title('Tổng tiền tip theo thời gian dùng bữa')
plt.xlabel('Thời gian dùng bữa')
plt.ylabel('Tổng tiền tip')
plt.show()
````

### 3. Vẽ đồ thị histogram của tổng hoá đơn

````python
# Vẽ đồ thị histogram của tổng hóa đơn
plt.figure(figsize=(8,6))
sns.histplot(tips_data['total_bill'], bins=20, kde=True)
plt.title('Phân phối tổng hóa đơn')
plt.xlabel('Tổng hóa đơn')
plt.ylabel('Tần số')
plt.show()
````

### 4. Vẽ đồ thị boxplot của tổng hoá đơn


````python
# Vẽ đồ thị boxplot của tổng hóa đơn
plt.figure(figsize=(8,6))
sns.boxplot(x='total_bill', data=tips_data)
plt.title('Boxplot của tổng hóa đơn')
plt.xlabel('Tổng hóa đơn')
plt.show()
````

### 5. Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần


````python
# Vẽ đồ thị boxplot so sánh tổng hóa đơn giữa các ngày trong tuần
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='total_bill', data=tips_data, order=['Thur', 'Fri', 'Sat', 'Sun'])
plt.title('Boxplot tổng hóa đơn theo ngày trong tuần')
plt.xlabel('Ngày trong tuần')
plt.ylabel('Tổng hóa đơn')
plt.show()
````

### 6. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip. (`scatter plot`)


````python
# Vẽ đồ thị scatter plot kiểm tra sự liên quan giữa tổng hóa đơn và tiền tip
sns.scatterplot(x='total_bill', y='tip', data=tips_data)
plt.title('Scatter Plot: Tổng hóa đơn vs Tiền tip')
plt.xlabel('Tổng hóa đơn')
plt.ylabel('Tiền tip')
plt.show()
````

### 7. Tìm hiểu và sử dụng hàm `jointplot` trong seaborn để khảo sát quan hệ giữa hai thuộc tính `total_bill` và `tip`. Theo bạn hai thuộc tính này có tương quan với nhau hay không?


````python
# Sử dụng jointplot để khảo sát quan hệ giữa total_bill và tip
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='scatter')
plt.show()  
````

**Nhận xét:**

- Biểu đồ phân phối của biến total_bill cho thấy hầu hết các hóa đơn có giá trị từ 10 đến 20 đô la, với trung bình khoảng 19.76 đô la. Biểu đồ có hình dạng lệch phải, tức là có nhiều hóa đơn có giá trị thấp hơn trung bình hơn là cao hơn trung bình.

- Biểu đồ phân phối của biến tip cho thấy hầu hết các lần boa có số tiền từ 2 đến 4 đô la, với trung bình khoảng 3 đô la. Biểu đồ cũng có hình dạng lệch phải, tức là có nhiều lần boa có số tiền thấp hơn trung bình hơn là cao hơn trung bình.

- Biểu đồ phân phối của hai biến total_bill và tip cho thấy có một mối tương quan dương khá rõ ràng giữa chúng, tức là khi tổng số tiền thanh toán càng cao thì số tiền boa cũng càng cao. Tuy nhiên, mối tương quan này không phải là tuyến tính hoàn hảo, tức là không phải cứ tăng 1 đơn vị total_bill thì tip cũng tăng 1 đơn vị tương ứng. Có thể thấy rằng có một số điểm ngoại lai (outliers) nằm xa so với phần lớn các điểm khác, ví dụ như có một hóa đơn có giá trị khoảng 50 đô la nhưng chỉ được boa 5 đô la, hoặc có một hóa đơn có giá trị khoảng 10 đô la nhưng lại được boa 6 đô la.

### 9. Vẽ hai đồ thị histogram cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`



````python
# Tạo subplot với 1 dòng và 2 cột
plt.figure(figsize=(12, 6))

# Vẽ histogram cho bữa trưa
plt.subplot(1, 2, 1)
sns.histplot(tips_data[tips_data['time'] == 'Lunch']['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Tổng hóa đơn - Lunch')
plt.xlabel('Tổng hóa đơn')
plt.ylabel('Số lượng')

# Vẽ histogram cho bữa tối
plt.subplot(1, 2, 2)
sns.histplot(tips_data[tips_data['time'] == 'Dinner']['total_bill'], bins=20, kde=True, color='salmon')
plt.title('Tổng hóa đơn - Dinner')
plt.xlabel('Tổng hóa đơn')
plt.ylabel('Số lượng')

# Điều chỉnh khoảng cách giữa các subplot
plt.subplots_adjust(wspace=0.3)

# Hiển thị đồ thị
plt.show()
````
