### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:

a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo quần thể mô phỏng
np.random.seed(0)
POP = np.random.uniform(120, 200, size=10000)
print("Bài 1: ")

# a. Chiều cao trung bình và độ lệch chuẩn
mean_height = np.mean(POP)
std_height = np.std(POP)
print("\na. Chiều cao trung bình và độ lệch chuẩn")
print("Chiều cao trung bình của quần thể:", mean_height)
print("Độ lệch chuẩn về chiều cao của quần thể:", std_height)

# b. Tỷ lệ người cao trong quần thể
tall_count = np.sum(POP >= 180)
tall_ratio = tall_count / len(POP)
print("\nb. Tỷ lệ người cao trong quần thể")
print("Tỷ lệ người cao trong quần thể:", tall_ratio)

# c. Vẽ histogram chiều cao của quần thể
print("\nc. Vẽ histogram chiều cao của quần thể")
plt.hist(POP, bins=50, edgecolor='black')
plt.xlabel('Chiều cao')
plt.ylabel('Số lượng')
plt.title('Phân phối chiều cao của quần thể')
plt.show()

# d. Vẽ histogram và đồ thị hàm mật độ phân phối chuẩn
print("\nd. Vẽ histogram và đồ thị hàm mật độ phân phối chuẩn")
x = np.linspace(120, 200, 1000)
y = (1 / (std_height * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_height) / std_height) ** 2)

plt.hist(POP, bins=50, density=True, alpha=0.5, edgecolor='black', label='Quần thể')
plt.plot(x, y, color='red', label='Phân phối chuẩn')
plt.xlabel('Chiều cao')
plt.ylabel('Mật độ xác suất')
plt.title('Phân phối chiều cao của quần thể và phân phối chuẩn')
plt.legend()
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
import numpy as np
import matplotlib.pyplot as plt

print("Bài 4: ")

# a. Thực hiện bài 2 100 lần và vẽ histogram cho 100 trung bình mẫu
print("\na. Thực hiện bài 2 100 lần và vẽ histogram cho 100 trung bình mẫu")
np.random.seed(0)
sample_means_20 = []

for _ in range(100):
    sample = np.random.choice(POP, size=20)
    sample_mean = np.mean(sample)
    sample_means_20.append(sample_mean)

plt.hist(sample_means_20, bins=10, edgecolor='black')
plt.xlabel('Trung bình mẫu')
plt.ylabel('Số lượng')
plt.title('Phân phối của trung bình mẫu (kích thước mẫu 20) - 100 lần')
plt.show()

# b. Thực hiện bài 3 100 lần và vẽ histogram cho 100 trung bình mẫu
print("\nb. Thực hiện bài 3 100 lần và vẽ histogram cho 100 trung bình mẫu")

sample_means_100 = []

for _ in range(100):
    sample = np.random.choice(POP, size=100)
    sample_mean = np.mean(sample)
    sample_means_100.append(sample_mean)

plt.hist(sample_means_100, bins=10, edgecolor='black')
plt.xlabel('Trung bình mẫu')
plt.ylabel('Số lượng')
plt.title('Phân phối của trung bình mẫu (kích thước mẫu 100) - 100 lần')
plt.show()

# c. Vẽ histogram của cả hai câu a và b để so sánh
print("\nc. Vẽ histogram của cả hai câu a và b để so sánh")
plt.hist(sample_means_20, bins=10, alpha=0.5, color='blue', edgecolor='black', label='kích thước mẫu 20')
plt.hist(sample_means_100, bins=10, alpha=0.5, color='red', edgecolor='black', label='kích thước mẫu 100')
plt.xlabel('Trung bình mẫu')
plt.ylabel('Số lượng')
plt.title('So sánh phân phối của trung bình mẫu')
plt.legend()
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
import numpy as np
import matplotlib.pyplot as plt

# Mô phỏng 1000 lần lấy mẫu kích thước 100 và vẽ histogram đại lượng phương sai của các mẫu
np.random.seed(0)
variances = []

print("Bài 6: Mô phỏng 1000 lần lấy mẫu kích thước 100 và vẽ histogram đại lượng phương sai của các mẫu")

for _ in range(1000):
    sample = np.random.choice(POP, size=100)
    variance = np.var(sample)
    variances.append(variance)

plt.hist(variances, bins=10, edgecolor='black')
plt.xlabel('Đại lượng phương sai')
plt.ylabel('Số lượng')
plt.title('Phân phối của đại lượng phương sai (kích thước mẫu 100) - 1000 lần')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
from scipy import stats

print("Bài 7: ")

# Dữ liệu cân nặng gà
weights = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])

# a. Khoảng tin cậy khi biết σ = 0.3
print("\na. Khoảng tin cậy khi biết σ = 0.3")
confidence_level = 0.95
sigma_known = 0.3

# Số lượng mẫu
n = len(weights)

# Tính khoảng tin cậy
mean = np.mean(weights)
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (sigma_known / np.sqrt(n))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print("Khoảng tin cậy khi biết σ = 0.3:")
print("Trung bình mẫu: {:.2f}".format(mean))
print("Khoảng tin cậy: ({:.2f}, {:.2f})".format(confidence_interval[0], confidence_interval[1]))
print()

# b. Khoảng tin cậy khi không biết σ
print("\nb. Khoảng tin cậy khi không biết σ")

confidence_level = 0.95

# Tính khoảng tin cậy
mean = np.mean(weights)
std_dev = np.std(weights, ddof=1)
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=n-1) * (std_dev / np.sqrt(n))
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print("Khoảng tin cậy khi không biết σ:")
print("Trung bình mẫu: {:.2f}".format(mean))
print("Khoảng tin cậy: ({:.2f}, {:.2f})".format(confidence_interval[0], confidence_interval[1]))
```
