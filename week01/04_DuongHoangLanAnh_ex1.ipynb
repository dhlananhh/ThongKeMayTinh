9. Viết hàm đưa vào 1 list số nguyên L và 1 số nguyên dương a. Hãy tính và trả về giá trị trung bình của a phần tử đầu tiên trong L

````python
# Viết hàm tính trung bình cộng của a phần tử đầu tiên trong một list
def average_of_first_a_elements(L, a):
    # Trả về None nếu a lớn hơn độ dài của danh sách
    if a <= len(L):
        return sum(L[:a]) / a
    # Ngược lại trả về trung bình cộng của a phần tử đầu tiên
    else:
        return None

L = [1, 2, 3, 4, 5, 6, 7, 8, 9]
a = 3
result = average_of_first_a_elements(L, a)

if result is not None:
    print(f"Giá trị trung bình của phần tử {a} đầu tiên trong danh sách là: {result}")
else:
    print(f"Số phần tử {a} trong danh sách không đủ để tính trung bình.")
````

11. Viết hàm có tham số đầu vào là một list L có các phần tử là chuỗi. Hãy tạo ra một dictionary D mã hóa, với mỗi một phần tử trong L được mã hóa thành một con số (theo thứ tự từ 0 tăng dần lên 1 đơn vị). Sau đó trả về list đã được mã hóa

    Ví dụ:

    ```
    Cho L = ["đen","vàng","xanh","vàng","xanh","đỏ","hồng"]

    Xây dựng dictionary mã hóa:
    D = {'đen':0,'vàng':1,'xanh':2,'đỏ':3,'hồng':4}

    Trả về List mã hóa:
    L_mahoa = [0, 1, 2, 1, 2, 3, 4]
    ```

````python
def encode_list(L):
    D = {v: k for k, v in enumerate(sorted(set(L)))}
    L_encoded = [D[i] for i in L]
    return L_encoded

L = ["đen", "vàng", "xanh", "vàng", "xanh", "đỏ", "hồng"]
print(encode_list(L))
````

12. Viết hàm kết hợp hai dictionary dựa vào key như sau:

    ```python
    d1 = {'a': 100, 'b': 200, 'c':300}
    d2 = {'a': 300, 'b': 200, 'd':400}
    Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})
    ```

````python
from collections import Counter

def combine_dicts(d1, d2):
    return dict(Counter(d1) + Counter(d2))

d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
print(combine_dicts(d1, d2))
````

15. Phát sinh ngẫu nhiên một mảng ndarray có kích thước là n nhập từ phím, mỗi phần tử là các số nguyên có giá trị từ 1..100. Bạn hãy thực hiện:
- Tính tổng mảng
- Tìm phần tử lớn nhất
- Tìm giá trị trung bình, trung vị, phân vị 50 và phân vị 75
- Tìm miền giá trị, miền phân vị
- Tính phương sai và độ lệch chuẩn
- Có bao nhiêu phần tử có một chữ số?
- In ra các phần tử có giá trị từ 80..100. Tính tổng các phần tử này.
- In ra m giá trị lớn nhất trong mảng. Biết rằng m = n//4
- Bạn hãy tạo ra một mảng từ mảng ban đầu; biết rằng mỗi phần tử của mảng mới bằng một phần tử của mảng ban đầu chia lấy dư cho 20.
- Sắp xếp mảng giảm dần.

````python
import numpy as np

# Phát sinh mảng ngẫu nhiên
n = int(input("Nhập kích thước mảng n: "))
arr = np.random.randint(1, 101, size=n)

# Tính tổng mảng
tong_mang = np.sum(arr)

# Tìm phần tử lớn nhất
max_element = np.max(arr)

# Tìm giá trị trung bình, trung vị, phân vị 50 và phân vị 75
trung_binh = np.mean(arr)
trung_vi = np.median(arr)
phan_vi_50 = np.percentile(arr, 50)
phan_vi_75 = np.percentile(arr, 75)

# Tìm miền giá trị, miền phân vị
mien_gia_tri = np.ptp(arr)
mien_phan_vi = np.percentile(arr, 75) - np.percentile(arr, 25)

# Tính phương sai và độ lệch chuẩn
phuong_sai = np.var(arr)
do_lech_chuan = np.std(arr)

# Số lượng phần tử có một chữ số
so_luong_mot_chu_so = np.sum((arr >= 1) & (arr <= 9))

# In ra các phần tử có giá trị từ 80 đến 100 và tính tổng
phane_tu_80_100 = arr[(arr >= 80) & (arr <= 100)]
tong_phan_tu_80_100 = np.sum(phane_tu_80_100)

# In ra m giá trị lớn nhất trong mảng (m = n//4)
m = n // 4
m_gia_tri_lon_nhat = np.sort(arr)[-m:]

# Tạo một mảng mới từ mảng ban đầu
mang_moi = arr % 20

# Sắp xếp mảng giảm dần
mang_giam_dan = np.sort(arr)[::-1]

# In kết quả
print("Mảng ngẫu nhiên:", arr)
print("Tổng mảng:", tong_mang)
print("Phần tử lớn nhất:", max_element)
print("Giá trị trung bình:", trung_binh)
print("Trung vị:", trung_vi)
print("Phân vị 50:", phan_vi_50)
print("Phân vị 75:", phan_vi_75)
print("Miền giá trị:", mien_gia_tri)
print("Miền phân vị:", mien_phan_vi)
print("Phương sai:", phuong_sai)
print("Độ lệch chuẩn:", do_lech_chuan)
print("Số lượng phần tử có một chữ số:", so_luong_mot_chu_so)
print("Phần tử từ 80 đến 100:", phane_tu_80_100)
print("Tổng các phần tử từ 80 đến 100:", tong_phan_tu_80_100)
print(f"{m} giá trị lớn nhất trong mảng:", m_gia_tri_lon_nhat)
print("Mảng mới từ mảng ban đầu (mỗi phần tử chia lấy dư cho 20):", mang_moi)
print("Mảng giảm dần:", mang_giam_dan)
````

16. Phát sinh ngẫu nhiên mảng các số nguyên có 10 phần tử có giá trị từ 1..100. Bạn hãy viết chương trình chọn ngẫu nhiên n mẫu (có hoàn lại), mỗi mẫu gồm 4 phần tử từ mảng ban đầu (với n nhập từ bàn phím). Thực hiện:

- Hiển thị các mẫu trên
- Tính tổng các mẫu trên
- Hiển thị mảng gồm n phần tử, mỗi phần tử chứa giá trị trung bình của các mẫu trên


````python
# Phát sinh mảng ngẫu nhiên có 10 phần tử
mang_ngau_nhien = np.random.randint(1, 101, size=10)

# Nhập số lượng mẫu n
n = int(input("Nhập số lượng mẫu n: "))

# Phát sinh mẫu ngẫu nhiên
mau_ngau_nhien = np.random.choice(mang_ngau_nhien, size=(n, 4), replace=True)

# Hiển thị các mẫu
print("Các mẫu:")
print(mau_ngau_nhien)

# Tính tổng các mẫu
tong_mau = np.sum(mau_ngau_nhien, axis=1)
print("\nTổng mẫu:")
print(tong_mau)

# Hiển thị mảng gồm n phần tử, mỗi phần tử chứa giá trị trung bình của các mẫu
trung_binh_mau = np.mean(mau_ngau_nhien, axis=1)
print("\nMảng gồm n phần tử, mỗi phần tử chứa giá trị trung bình của các mẫu:")
print(trung_binh_mau)
````

17. Phát sinh ngẫu nhiên mảng các số nguyên có 10 phần tử có giá trị từ 1..100. Bạn hãy viết chương trình hiển thị 10 hoán vị không giống nhau và không giống với mỗi ban đầu.

````python
# Phát sinh mảng ngẫu nhiên có 10 phần tử
mang_ngau_nhien = np.random.randint(1, 101, size=10)

# Hiển thị mảng ban đầu
print("Mảng ban đầu:", mang_ngau_nhien)

# Phát sinh và hiển thị 10 hoán vị không giống nhau và không giống với mảng ban đầu
print("\n10 hoán vị không giống nhau:")
for _ in range(10):
    hoan_vi = np.random.permutation(mang_ngau_nhien)
    print(hoan_vi)
````
