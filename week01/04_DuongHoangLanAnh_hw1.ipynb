2. Viết chương trình nhập vào số nguyên n có 3 chữ số. Tính tổng các chữ số của n

```python
# Nhập số nguyên n có 3 chữ số
n = int(input("Nhập n: "))

# Kiểm tra xem n có phải là số nguyên dương có 3 chữ số hay không
if 100 <= n <= 999:
    # Tính tổng các chữ số
    hang_tram = n // 100
    hang_chuc = (n % 100) // 10
    hang_don_vi = n % 10

    tong_chu_so = hang_tram + hang_chuc + hang_don_vi

# In kết quả
    print(f"Tổng các chữ số của {n} là: {tong_chu_so}")
else:
    print("Vui lòng nhập vào số nguyên dương có 3 chữ số")
```

4. Viết chương trình yêu cầu người dùng nhập họ tên. Hiển thị tên người dùng sau khi đã chuẩn hóa. Biết rằng tên đã chuẩn hóa là:
- Ký tự đầu viết hoa
- Không có khoảng trắng thừa
- Không có ký tự đặc biệt

```python
# Tạo hàm chuẩn hóa tên người dùng
def standardize_name(name):
    # Chuyển ký tự đầu của mỗi từ thành chữ hoa
    standardizeName = name.title()
    
    # Loại bỏ khoảng trắng thừa ở đầu và cuối chuỗi
    standardizeName = name.strip()
    
    # Loại bỏ ký tự đặc biệt
    standardizeName = ''.join(char for char in standardizeName if char.isalpha() or char.isspace())
    
    # Trả về kết quả là tên đã được chuẩn hóa
    return standardizeName


# Yêu cầu người dùng nhập họ tên
user_input = input("Nhập họ tên của bạn: ")


# Chuẩn hóa và hiển thị tên
standardized_name = standardize_name(user_input)
print("Tên đã chuẩn hóa:", standardized_name)
```

5. Viết chương trình nhập vào một số nguyên dương (n>0). Cho biết n có phải là số đối xứng hay không? Biết số đối xứng là số viết xuôi hay viết ngược đều như nhau

```python
# Tạo hàm kiểm tra xem n có phải là số đối xứng hay không
def isPalindrome (number):
    # Chuyển số thành chuỗi để dễ xử lý
    str_number = str(number)
    
    # So sánh chuỗi với chuỗi đảo ngược
    return str_number == str_number[::-1]


# Yêu cầu người dùng nhập số nguyên dương n
n = int(input("Nhập số nguyên dương n (n > 0): "))


# Kiểm tra và hiển thị kết quả
if (n > 0):
    if isPalindrome(n):
        print(f"{n} là số đối xứng")
    else:
        print(f"{n} không là số đối xứng")
else:
    print("Vui lòng nhập lại số nguyên dương n!")
```

6. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
import numpy as np


# Phát sinh mảng ngẫu nhiên
arr = np.random.randint(0, 10001, size=100)

'''
Lọc các phần tử của mảng 
thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100
'''
filtered_array = [x for x in arr if x % 5 == 0 and x > 100]


# In kết quả
print("Mảng vừa phát sinh ngẫu nhiên: \n", arr)
print("\n10 phần tử đầu tiên của mảng: \n", arr[:10])
print("\n5 phần tử cuối cùng của mảng: \n", arr[-5:])
print("\nCác phần tử của mảng thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100 là: \n", filtered_array)
```
