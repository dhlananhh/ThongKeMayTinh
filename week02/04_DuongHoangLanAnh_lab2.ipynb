### Câu 6: Vẽ đồ thị cột

### Câu  7: Vẽ đồ thị bánh

```python
import matplotlib.pyplot as plt

# Dữ liệu
languages = ['C#', 'C++', 'Java', 'Python', 'PHP', 'JavaScript']
percentages = [10.85, 9.44, 31.27, 24.79, 12.39, 11.27]
explode = (0, 0, 0.1, 0, 0, 0)  # Phần bánh Java rời ra

# Màu sắc pastel
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700', '#FF6666']

# Vẽ pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=languages, autopct='%1.2f%%', startangle=0, explode=explode, colors=colors)
plt.title('Mức độ phổ biến của các ngôn ngữ lập trình')
plt.axis('equal')  # Đảm bảo pie chart là hình tròn

# Thêm legend
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Hiển thị pie chart
plt.show()
```

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
# Your code here

import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên thuộc phân phối chuẩn
random_data = np.random.normal(size=10000)

# Vẽ histogram
plt.hist(random_data, bins=30, color='#86bf91', edgecolor='black')

# Thêm label cho trục và tiêu đề
plt.xlabel('Giá trị')
plt.ylabel('Số lượng')
plt.title('Histogram của 10000 số ngẫu nhiên phân phối chuẩn')

# Hiển thị biểu đồ
plt.show()
```

### Câu 9: Vẽ đồ thị hộp

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ boxplot các số ngẫu nhiên trên.

```python
# Your code here

import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên thuộc phân phối chuẩn
random_data = np.random.normal(size=10000)

# Vẽ boxplot
plt.boxplot(random_data)

# Thêm label cho trục và tiêu đề
plt.xlabel('Dãy số ngẫu nhiên')
plt.ylabel('Giá trị')
plt.title('Boxplot của 10000 số ngẫu nhiên phân phối chuẩn')

# Hiển thị biểu đồ
plt.show()
```

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

```python
# Your code here

import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên thuộc phân phối chuẩn
random_data = np.random.normal(size=10000)

# Tạo figure và axes
fig, (ax1, ax2) = plt.subplots(2, sharex=True, gridspec_kw={'height_ratios': [3, 1]})

# Vẽ histogram ở phần trên
ax1.hist(random_data, bins=30, color='#86bf91', edgecolor='black')
ax1.set_ylabel('Số lượng')
ax1.set_title('Histogram và Boxplot của 10000 số ngẫu nhiên phân phối chuẩn')

# Vẽ boxplot ở phần dưới
ax2.boxplot(random_data, vert=False)
ax2.set_xlabel('Giá trị')
ax2.set_ylabel('')

# Hiển thị biểu đồ
plt.show()
```
