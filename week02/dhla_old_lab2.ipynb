```python
import numpy as np
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt


window = signal.windows.triang(51)
plt.plot(window)
plt.title("Triangular window")
plt.ylabel("Amplitude")
plt.xlabel("Sample")


plt.figure()
A = fft(window, 2048) / (len(window)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = np.abs(fftshift(A / abs(A).max()))
response = 20 * np.log10(np.maximum(response, 1e-10))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Frequency response of the triangular window")
plt.ylabel("Normalized magnitude [dB]")
plt.xlabel("Normalized frequency [cycles per sample]")
```

```python
import matplotlib.pyplot as plt

# Dữ liệu
languages = ['C#', 'C++', 'Java', 'Python', 'PHP', 'JavaScript']
percentages = [10.85, 9.44, 31.27, 24.79, 12.39, 11.27]
explode = (0, 0, 0.1, 0, 0, 0)  # Phần bánh Java rời ra

# Màu sắc pastel
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700', '#FF6666']

# Vẽ pie chart
plt.figure(figsize=(8, 8))
plt.pie(percentages, labels=languages, autopct='%1.2f%%', startangle=45, explode=explode, colors=colors)
plt.title('Mức độ phổ biến của các ngôn ngữ lập trình')
plt.axis('equal')  # Đảm bảo pie chart là hình tròn

# Thêm legend
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Hiển thị pie chart
plt.show()
```

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

```python
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên thuộc phân phối chuẩn
random_data = np.random.normal(size=10000)

# Vẽ histogram
plt.hist(random_data, bins=30, color='#86bf91', edgecolor='black')

# Thêm label cho trục và tiêu đề
plt.xlabel('Giá trị')
plt.ylabel('Số lượng')
plt.title('Histogram của 10000 số ngẫu nhiên phân phối chuẩn')

# Hiển thị biểu đồ
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên thuộc phân phối chuẩn
random_data = np.random.normal(size=10000)

# Vẽ boxplot
plt.boxplot(random_data)

# Thêm label cho trục và tiêu đề
plt.xlabel('Dãy số ngẫu nhiên')
plt.ylabel('Giá trị')
plt.title('Boxplot của 10000 số ngẫu nhiên phân phối chuẩn')

# Hiển thị biểu đồ
plt.show()
```
