Câu 2 (1 điểm): Tóm tắt dữ liệu cột "Chiều cao" bằng kỹ thuật Thân/lá (stem/leaf plot)

```python
height_data = df['Chiều Cao'].to_dict()
```

```python
print("Dữ liệu chiều cao dưới dạng dictionary: ")
print(height_data)
```

```python
# Hàm để tạo biểu đồ Thân/lá
def stem_and_leaf(data):
    stems = [int(x // 10) for x in data]
    leaves = [int(x % 10) for x in data]

    stem_leaf_dict = {s: [] for s in stems}
    for s, l in zip(stems, leaves):
        stem_leaf_dict[s].append(l)

    return stem_leaf_dict

# Tính toán và in biểu đồ Thân/lá
stem_leaf_data = stem_and_leaf(height_data)

print("Stem-and-Leaf Plot:")
for stem, leaves in sorted(stem_leaf_data.items()):
    print(f"{stem} | {' '.join(map(str, sorted(leaves)))}")

# Vẽ biểu đồ Thân/lá bằng matplotlib
fig, ax = plt.subplots()
for stem, leaves in stem_leaf_data.items():
    values = [stem * 10 + leaf for leaf in leaves]
    ax.scatter([stem] * len(values), values, marker='o')

ax.set_xlabel('Stem')
ax.set_ylabel('Leaves')
ax.set_title('Stem-and-Leaf Plot for Height')

plt.show()
```

Câu 3 (1 điểm): Tóm tắt dữ liệu cột "Cân nặng" bằng bảng phân phối tần số với độ rộng khoảng bằng 10

```python
weight_data = df['Cân Nặng'].to_dict()
```

```python
print("Dữ liệu cân nặng dưới dạng dictionary: ")
print(weight_data)
```

```python
# Vẽ biểu đồ histogram
plt.hist(df['Cân Nặng'], bins=range(0, int(max(df['Cân Nặng'])) + 10, 10), edgecolor='black')
plt.xlabel('Cân nặng')
plt.ylabel('Số lượng')
plt.title('Biểu đồ Histogram cho Cân nặng')
plt.grid(True)
plt.show()
```

```python
# Tính toán bảng phân phối tần số
bin_width = 10
weight_bins = [i for i in range(0, int(df['Cân Nặng'].max()) + bin_width, bin_width)]
frequency_table = pd.cut(df['Cân Nặng'], bins=weight_bins, right=False).value_counts().sort_index()

# In bảng phân phối tần số
print("Bảng Phân Phối Tần Số của Cân Nặng:")
print(frequency_table)
```

```python
# Thiết lập độ rộng khoảng
bin_width = 50

# Vẽ Histogram
plt.hist(df['Toeic'], bins=range(0, int(df['Toeic'].max()) + bin_width, bin_width), edgecolor='black')

# Đặt các nhãn và tiêu đề
plt.xlabel('Điểm Toeic')
plt.ylabel('Số Lượng Sinh Viên')
plt.title('Histogram của Điểm Toeic')

# Hiển thị đồ thị
plt.show()
```
