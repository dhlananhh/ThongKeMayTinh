import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

### Bài 1

Nghiên cứu mức thu nhập ($X$) và chi tiêu ($Y$) trong một tháng của một công ty. Khảo sát ngẫu nhiên 7 nhân viên, ta thu được bảng số liệu sau:

| X            | 130     | 150      | 133      | 170      | 170      | 210      | 230      |
| :----------: | :-----: | :------: | :------: | :------: | :------: | :------: | :------: |
| **Y**        | **84**  | **120**  | **108**  | **130**  | **130**  | **150**  | **160**  |


a. Dựa vào bảng dữ liệu trên cho biết có xây dựng được mô hình hồi quy hay không? Nếu có hãy xây dựng mô hình hồi quy để ước tính chi tiêu dựa theo thu nhập

b. Dự đoán mức chi tiêu của nhân viên công ty trên nếu biết thu nhập là 200$

```python
# Chuẩn bị dữ liệu

# Tạo dataframe
df1 = pd.DataFrame({
    "Thu nhập": [130, 150, 133, 170, 170, 210, 230],
    "Chi tiêu": [84, 120, 108, 130, 130, 150, 160]
})

# Chuyển đổi dữ liệu sang dạng mảng
X1 = df1["Thu nhập"].to_numpy().reshape(-1, 1)
y1 = df1["Chi tiêu"].to_numpy()
```

```python
# Huấn luyện mô hình

# Khởi tạo mô hình hồi quy tuyến tính
model1 = LinearRegression()

# Huấn luyện mô hình với dữ liệu
model1.fit(X1, y1)
```

```python
# Hiển thị thông tin mô hình

# Hệ số hồi quy
print(f"Hệ số a: {model1.intercept_}")
print(f"Hệ số b: {model1.coef_[0]}")

# R^2
print(f"R^2: {model1.score(X1, y1)}")
```

```python
# Dự đoán mức chi tiêu
income = np.array([200]).reshape((-1, 1))
predicted_expense = model1.predict(income)

print("Mô hình hồi quy: chi tiêu = {:.2f} * thu nhập".format(model1.coef_[0]))
print("Dự đoán mức chi tiêu với thu nhập là 200: {:.2f}".format(predicted_expense[0]))
```

```python
# Vẽ đồ thị

plt.scatter(X1, y1)
plt.plot(X1, model1.predict(X1), color="red")
plt.xlabel("Thu nhập")
plt.ylabel("Chi tiêu")
plt.show()
```

### Bài 2

Tiến hành nghiên cứu mối quan hệ giữa:

- X: giá bán kẻ của thịt gà (nghìn/kg) 
- Y: lượng thịt gà tiêu thụ (kg/tháng)

Dữ liệu thu được gồm 10 tháng như sau:

| X            | 35       | 40       | 45       | 40       | 38       | 40       | 45       | 30       | 42       | 38       |
| :----------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: |
| **Y**        | **2.29** | **2.04** | **0.59** | **2.06** | **2.50** | **2.32** | **1.57** | **3.96** | **1.43** | **2.34** |


a. Dựa vào bảng dữ liệu trên cho biết có xây dựng được mô hình hồi quy hay không? Nếu có hãy xây dựng mô hình hồi quy để ước tính lượng tiêu thụ thịt gà theo giá bán.

b. Dự đoán lượng tiêu thụ khi giá bán lẻ là 41(kg/tháng)

```python
# Chuẩn bị dữ liệu

# Tạo dataframe
df2 = pd.DataFrame({
    "Giá bán": [35, 40, 45, 40, 38, 40, 45, 30, 42, 38],
    "Lượng tiêu thụ": [2.29, 2.04, 0.59, 2.06, 2.50, 2.32, 1.57, 3.96, 1.43, 2.34]
})

# Chuyển đổi dữ liệu sang dạng mảng
X2 = df2["Giá bán"].to_numpy().reshape(-1, 1)
y2 = df2["Lượng tiêu thụ"].to_numpy()
```

```python
# Huấn luyện mô hình

# Khởi tạo mô hình hồi quy tuyến tính
model2 = LinearRegression()

# Huấn luyện mô hình với dữ liệu
model2.fit(X2, y2)
```

```python
# Hiển thị thông tin mô hình

# Hệ số hồi quy
print(f"Hệ số a: {model2.intercept_}")
print(f"Hệ số b: {model2.coef_[0]}")

# R^2
print(f"R^2: {model2.score(X2, y2)}")
```

```python
# Dự đoán lượng tiêu thụ 

retail_price = np.array([41]).reshape((-1, 1))
predicted_consumption = model2.predict(retail_price)

print("Mô hình hồi quy: lượng tiêu thụ = {:.2f} * giá bán lẻ".format(model2.coef_[0]))
print("Dự đoán lượng tiêu thụ với giá bán lẻ là 41 kg/tháng: {:.2f}".format(predicted_consumption[0]))
```

```python
# Vẽ đồ thị

plt.scatter(X2, y2)
plt.plot(X2, model2.predict(X2), color="red")
plt.xlabel("Giá bán (nghìn/kg)")
plt.ylabel("Lượng tiêu thụ (kg/tháng)")
plt.show()
```

## Bài 3

Dụa trên file dữ liệu `Advertising.csv` bạn hãy dự đoán doanh số bán hàng (`sales`) dựa trên số tiền chi cho các nền tảng tiếp thị khác nhau như `TV`, `Radio` và `Newspaper`

a. Đọc file dữ liệu cho biết các thông tin cơ bản về dữ liệu: kích thước, dữ liệu có giá trị trống hay không, hiện các giá trị thống kê cơ bản của các thuộc tính.

b. Để dự đoán doanh số bằng phương pháp hồi quy tuyến tính đơn giản (Simple Linear Regression) trước tiên cần lựa chọn một trong các thuộc tính còn lại. Theo bạn thuộc tính nào phù hợp để xây dựng mô hình hồi quy. Giải thích và minh họa bằng 2 cách

c. Xây dụng phương trình hồi quy, trực quan mô hình (đường thẳng) tìm được. Theo bạn để đánh giá mô hình trên cần dùng đại lượng nào?

### Câu a: Tìm hiểu dữ liệu

```python
# Đọc file dữ liệu
df3 = pd.read_csv("Advertising.csv")
```

```python
# Hiển thị dữ liệu
df3.head()
```

```python
# Kích thước dữ liệu
print("Kích thước dữ liệu:", df3.shape)
```

```python
# Kiểm tra giá trị trống
print("Dữ liệu có giá trị trống:\n", df3.isnull().sum())
```

```python
# Thống kê cơ bản của các thuộc tính
print("Thống kê cơ bản:\n", df3.describe())
```

### Câu b: Lựa chọn thuộc tính phù hợp cho mô hình hồi quy

#### Cách 1: Phân tích hệ số tương quan

```python
# Hiển thị ma trận tương quan
corr = df3.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

**Nhận xét:**

Quan sát ma trận tương quan, ta thấy:

- `TV` có mối tương quan cao với `sales` (0.78)

- `Radio` có mối tương quan trung bình với `sales` (0.58)

- `Newspaper` có mối tương quan thấp với `sales` (0.23)

#### Cách 2: Vẽ scatter plot

```python
sns.regplot(x="TV", y="sales", data=df3, line_kws={"color": "red"})
plt.xlabel("TV")
plt.ylabel("Sales")
plt.title('Linear Regression - TV vs Sales')
plt.show()
```

```python
sns.regplot(x="radio", y="sales", data=df3, line_kws={"color": "red"})
plt.xlabel("Radio")
plt.ylabel("Sales")
plt.title('Linear Regression - Radio vs Sales')
plt.show()
```

```python
sns.regplot(x="newspaper", y="sales", data=df3, line_kws={"color": "red"})
plt.xlabel("Newspaper")
plt.ylabel("Sales")
plt.title('Linear Regression - Newspaper vs Sales')
plt.show()
```

**Nhận xét:**

Quan sát 3 đồ thị trên, ta thấy:

- Đồ thị 1: Mối liên hệ giữa doanh số bán hàng (Sales) và chi tiêu quảng cáo TV (TV) là tuyến tính rõ ràng nhất.

- Đồ thị 2: Mối liên hệ giữa doanh số bán hàng (Sales) và chi tiêu quảng cáo Radio (Radio) cũng tuyến tính nhưng không rõ ràng bằng đồ thị 1.

- Đồ thị 3: Mối liên hệ giữa doanh số bán hàng (Sales) và chi tiêu quảng cáo Newspaper (Newspaper) không rõ ràng.

#### Câu c: Xây dựng mô hình hồi quy tuyến tính, trực quan hóa và đánh giá mô hình

```python
# Xây dựng mô hình hồi quy đơn giản
# Lựa chọn thuộc tính TV
X = df3[['TV']]
y = df3['sales']

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X, y)

# Minh họa mô hình
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('Linear Regression - TV vs Sales')
plt.show()

# Hiển thị thông tin mô hình
print("Hệ số hồi quy:", model.coef_[0])
print("Hệ số chặn:", model.intercept_)
print("R^2:", model.score(X, y))
print(f"Phương trình hồi quy: Sales = {model.coef_[0]:.2f} + {model.intercept_:.2f} * TV")
```

c. Để đánh giá mô hình, chúng ta sử dụng đại lượng R-square. Đại lượng này cho biết phần trăm phương sai của biến phụ thuộc (doanh số bán hàng) được giải thích bởi mô hình. Giá trị R-square càng gần 1, mô hình càng tốt.

```python
# c. Đánh giá mô hình
# Đánh giá mô hình bằng đại lượng R-square
r2_score = model.score(X, y)
print("Đánh giá mô hình - R-square:", r2_score)
```
